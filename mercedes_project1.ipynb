```python
X=tr_df.drop(['y'],axis=1)
```

```python
X.head(5)
```

```python
y=tr_df["y"]
```

```python
from sklearn.decomposition import PCA
pca_actual=PCA(0.98,svd_solver='full')
pca = PCA(n_components = 2)
X1 = pca.fit_transform(X)
#X_test = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
X_actual=pca_actual.fit_transform(X)
print(pca_actual.explained_variance_ratio_)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X1,y,test_size=0.25)
x_train_12,x_test_12,y_train_12,y_test_12=train_test_split(X_actual,y,test_size=0.25)
from xgboost import XGBRegressor
classifier=XGBRegressor()
classifier_12=XGBRegressor()
```

```python
classifier.fit(x_train,y_train)
classifier_12.fit(x_train_12,y_train_12)
```

```python
print(classifier.feature_importances_)
classifier_12.feature_importances_
```

```python
y_pred=classifier.predict(x_test)
y_pred_12=classifier_12.predict(x_test_12)
```

```python
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_pred,y_test))
rmse_12=np.sqrt(mean_squared_error(y_pred_12,y_test_12))
```
